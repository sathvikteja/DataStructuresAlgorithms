```python
from collections import deque
class Node:
  def __init__(self,key):
    self.left=None
    self.right=None
    self.val=key
  def inorder(node):
    if node is None:
      return
    inorder(node.left)
    print(node.val,end=' ')
    inorder(node.right)
  def preorder(node):
    if node is None:
      return
    print(node.val,end='')
    self.preorder(node.left)
    self.preorder(node.right)

  def insert(root,key):
    new_node=Node(key)
    if not root:
      return new_node
    q=deque([root])
    while q:
      node=q.popleft()
      if not node.left:
        node.left=new_node
        return root
      else:
        q.append(node.left)
      if not node.right:
        node.right=new_node
        return root
      else:
        q.append(node.right)

```

```python
root = Node(2)
root.left = Node(3)
root.right = Node(4)
root.left.left = Node(5)
```
