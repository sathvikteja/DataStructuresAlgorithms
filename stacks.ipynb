```python
#Stacks
class Node:
  def __init__(self,value):
    self.data=value
    self.next=None
class Stack:
  def __init__(self):
    self.top=None

  def isempty(self):
    return self.top==None

  def push(self,value):
    new_node=Node(value)
    new_node.next=self.top
    self.top=new_node

  def peak(self):
    if self.isempty() :
      return "stack is empty"
    return self.top.data

  def traverse(self):
    curr=self.top
    if curr==None:
      return "empty stack"
    while(curr is not None):
      print(curr.data,end=' ')
      curr=curr.next

  def pop(self):
    if self.isempty() :
      return "stack is empty "
    t = self.top
    self.top=self.top.next
    return t.data

  def size(self):
    if self.isempty() :
      return "stack is empty"
    count=0
    curr=self.top
    while(curr is not None):
      curr=curr.next
      count+=1
    return count


s=Stack()
s.push(1)
s.push(2)
s.push(3)
s.traverse()
```

```python
def pattern(arr):
  st=[]
  t=-1
  max=-1000
  for i in range(len(arr)-1,-1,-1):
    while(st and st[t]>arr[i]):
      st.pop()
      t-=1
    st.append(arr[i])
    t+=1
    if max<st[t]:
      max=st[t]
    if len(st)>=2 and st[t]<max:
      return True
  return False



pattern([3,5,0.3,4])
```

```python
def reverse(text):
  s=Stack()
  for i in text:
    s.push(i)
    res=''
  while( not s.isempty()):
    res+=s.peak()
    s.pop()
  print(res)
reverse('hello')
```

```python
def twostack(text,cmd):
  s1=Stack()
  s2=Stack()

  for i in text:
    s1.push(i)

  for i in cmd:
    if i == 'u':
      data=s1.pop()
      s2.push(data)
    if i == 'r':
      if not s2.isempty():
        data=s2.pop()
        s1.push(data)

  res=''
  while(not s1.isempty()):
    t=s1.pop()
    res=t+res
  return res

twostack('hello','uuuuuu')
```
