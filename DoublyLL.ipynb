```python
# DOUBLY LINKED LISTS
class Node:
  def __init__(self,data):
    self.data=data
    self.prev=None
    self.next=None

class DoublyLL:
  def __init__(self):
    self.head=None

  def traverse(self):
    curr=self.head
    while(curr is not None):
      print(curr.data,end='<->')
      curr=curr.next
    print('None')

  def insert_at_beginning(self,data):
    new_node=Node(data)

    if self.head is None:
      self.head=new_node
      return
    new_node.next=self.head
    self.head.prev=new_node
    self.head=new_node

  def insert_after(self,node,data):
    if node is None:
      return "error"
    new_node=Node(data)
    new_node.next=node.next
    node.next=new_node
    new_node.prev=node

    if new_node.next is not None:
      new_node.next.prev=new_node

  def insert_before(self,node,data):
    if node is None:
      return "error"
    new_node=Node(data)
    new_node.prev=node.prev
    node.prev=new_node
    new_node.next=node
    if new_node.prev is not None:
      new_node.prev.next=new_node


  def insert_at_end(self,data):
    new_node=Node(data)
    curr=self.head
    if curr is None:
      return new_node
    while(curr.next is not None):
      curr=curr.next
    curr.next=new_node
    new_node.prev=curr
    new_node.next=None

# Deletion
  def delete_at_beginning(self):
    curr=self.head
    if curr is None:
      return "no nodes to delete"
    if curr.next is None:
      return "None"
    self.head=curr.next
    curr.prev=None
    del curr


  def delete_at_position(self,pos):
    if pos<0:
      return "invalid position"
    if pos==0:
      head=self.head
      if head.next:
        head.next.prev=None
      self.head=head.next
    if pos>0:
      count=0
      curr=self.head
      while(pos!=count):
        curr=curr.next
        count+=1
      curr.prev.next=curr.next
      curr.next.prev=curr.prev
  def delete_at_end(self):
    curr=self.head
    if curr is None:
      return "empty LL"
    if curr.next is None:
      self.head = None
      return None
    while(curr.next is not None):
      curr=curr.next
    curr.prev.next=None








l=DoublyLL()
l.insert_at_beginning(4)
l.insert_at_beginning(3)
l.insert_at_beginning(2)
l.insert_at_beginning(1)
l.traverse()
#l.insert_after(l.head.next,2.5)
#l.insert_before(l.head.next.next,2.5)
#l.insert_at_end(5)
#l.delete_at_beginning()
#l.delete_at_position(0)
#l.delete_at_end()
l.traverse()
```
